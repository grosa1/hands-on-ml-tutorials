### Image preprocessing
* **conversion to grayscale** (black and white): as these are X-rays, we are not interested in colors;
* **image resizing** with the size we defined in the variable IMG_SIZE;
* **normalization of the images** bringing all the pixels in a range from 0 to 1;
* **histogram equalization:** to limit contrast/brightness differences in the images;
* **coding of labels** assigning the normal case the value 0 (negative) and the covid case the value 1 (positive).

```
X = []
y = []
encoding = [("normal",0),("covid",1)] # define encoding
for folder, label in encoding:
  for img_name in os.listdir("dataset/"+folder):
    img = cv2.imread("dataset/"+folder+"/"+img_name, cv2.IMREAD_GRAYSCALE) # read the image in a np array
    img = cv2.equalizeHist(img)
    img = cv2.resize(img, IMG_SIZE)/255. # resize and normalization
    X.append(img)
    y.append(label)
X = np.array(X)
y = np.array(y)
```
